# Steam Games

```python
import seaborn as sns
import matplotlib.pyplot as plt

import pandas as pd
import os 

path_steam_games = os.path.join('..','data','clear','steam_games.csv.gz')

user_reviews = pd.read_csv('../data/clear/user_reviews.csv.gz',compression='gzip',lineterminator='\n')

user_items = pd.read_csv('../data/clear/users_items.csv.gz',compression='gzip')

steam_games = pd.read_csv(path_steam_games)
```

```python

data_filtered1 = steam_games[steam_games['release_year'].isin([2011,2012,2013,2014,2015,2016,2017,2018])]
data_filtered2 = user_reviews[user_reviews['year_posted'].isin([2011,2012,2013,2014,2015,2016,2017])]
data_filtered3 = steam_games['developer'][ steam_games['developer'] !='Otros'].value_counts().head(10)



plt.figure(figsize=(15, 14))
plt.subplot(3, 1, 1)
sns.countplot(x='release_year', data=data_filtered1, palette='viridis')
plt.title(f'Cantidad de juegos Lanzados por año')
plt.xlabel('Año de Lanzamiento')
plt.ylabel('Cantidad de juegos')
# Anotar cada barra con su valor
for barra in plt.gca().patches:
        plt.text(barra.get_x() + barra.get_width()/2, barra.get_height(), 
                 f'{int(barra.get_height())}', 
                 ha='center', va='bottom')

plt.subplot(3, 1, 2)
sns.countplot(x='year_posted', data=data_filtered2, palette='viridis')
plt.title(f'Cantidad de comentarios por año')
plt.xlabel('Año de publicación')
plt.ylabel('Cantidad de comentarios')
# Anotar cada barra con su valor
for barra in plt.gca().patches:
        plt.text(barra.get_x() + barra.get_width()/2, barra.get_height(), 
                 f'{int(barra.get_height())}', 
                 ha='center', va='bottom')
        
        
        
plt.subplot(3, 1, 3)
sns.barplot(x=data_filtered3.index, y=data_filtered3.values, palette='viridis')
plt.title(f'Cantidad de comentarios por año')
plt.xlabel('Año de publicación')
plt.ylabel('Cantidad de comentarios')
# Anotar cada barra con su valor
for barra in plt.gca().patches:
        plt.text(barra.get_x() + barra.get_width()/2, barra.get_height(), 
                 f'{int(barra.get_height())}', 
                 ha='center', va='bottom')
        
# Decorar el gráfico
plt.title(f'Top 10 de desarrolladores')
plt.xlabel('Desarrollador')
plt.ylabel('Cantidad de juegos lanzados')

# Rotar las etiquetas en el eje x para mayor legibilidad
plt.xticks(rotation=45, ha='right')


# Ajustar el diseño y mostrar el gráfico
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Supongamos que tienes un DataFrame llamado 'steam_games'
# steam_games = pd.read_csv('tu_archivo.csv')  # Asegúrate de cargar tus datos correctamente

# Calcular la proporción de juegos con y sin early access
early_access = steam_games['early_access'].value_counts(normalize=True)

# Crear un gráfico de barras horizontal
plt.figure(figsize=(12, 6))
sns.barplot(y=early_access.values, x=early_access.index, palette='viridis')

# Anotar las barras con sus valores
for barra in plt.gca().patches:
        plt.text(barra.get_x() + barra.get_width()/2, barra.get_height(), 
                 f'{round(barra.get_height(),2)}', 
                 ha='center', va='bottom')
# Decorar el gráfico
plt.title('Proporción de Juegos con Early Access')
plt.xlabel('Proporción de Juegos')
plt.ylabel('Early Access')

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Texto de ejemplo
texto_ejemplo = ' '.join(steam_games['developer'].str.split(', ').explode())

stop = ['Otros']

# Crear un objeto WordCloud
wordcloud = WordCloud(background_color = "white", max_words = 50,stopwords=stop).generate(texto_ejemplo)

# Visualizar la nube de palabras utilizando Matplotlib
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')  # No mostrar ejes
plt.show()
```

## Desarrollador que su juego se jugó en promedio más.

```python
desarrollador_playtime = user_items[['item_id','playtime_forever','item_name']]\
                                    .groupby(['item_id','item_name']).mean().reset_index()\
                                    .sort_values('playtime_forever',ascending=False).head(15)\
                                    .merge(steam_games[['id','developer']],
                                           left_on='item_id',
                                           right_on='id',
                                           how='inner')
                                    
```

```python
desarrollador_playtime
```

```python
plt.figure(figsize=(15, 14))

sns.barplot(x='developer', y='playtime_forever', data=desarrollador_playtime, palette='viridis')
plt.title(f'Cantidad de comentarios por año')
plt.xlabel('Año de publicación')
plt.ylabel('Cantidad de comentarios')
# Anotar cada barra con su valor
for barra in plt.gca().patches:
        plt.text(barra.get_x() + barra.get_width()/2, barra.get_height(), 
                 f'{int(barra.get_height())}', 
                 ha='center', va='bottom')
        
# Decorar el gráfico
plt.title(f'Top 10 de desarrolladores')
plt.xlabel('Desarrollador')
plt.ylabel('Tiempo promedio jugado')

# Rotar las etiquetas en el eje x para mayor legibilidad
plt.xticks(rotation=45, ha='right')
plt.show()
```
